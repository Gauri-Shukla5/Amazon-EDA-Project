# **Project Summary -**

This project focuses on performing Exploratory Data Analysis (EDA) on a dataset containing information about movies and TV shows, along with their cast and crew details. The dataset is divided into two parts: titles and credits. The titles dataset includes information such as movie/show name, type (movie or show), release year, genre, rating, and runtime. The credits dataset contains details about actors, directors, and other crew members associated with each title.

The main aim of this project is to understand patterns and trends in the entertainment industry by analyzing this data. With the help of EDA, we can explore how content has changed over time, which genres are most popular, and how ratings vary across different types of content. This helps in understanding audience preferences and content performance.

The project begins with loading both datasets and merging them using a common column (id). After merging, we perform data cleaning steps such as handling missing values, removing duplicate records, and correcting data types where necessary. Data cleaning is important to ensure that the analysis is accurate and reliable.

Next, we perform data visualization using different types of charts such as bar charts, histograms, scatter plots, box plots, heatmaps, and pair plots. These visualizations help us to better understand the distribution of variables and relationships between them. For example, we analyze how ratings are distributed, how the number of movies has changed over the years, and which genres are most common.

We also perform univariate, bivariate, and multivariate analysis. In univariate analysis, we study individual variables like rating or release year. In bivariate analysis, we explore relationships between two variables such as release year and rating, or genre and rating. In multivariate analysis, we analyze multiple variables together to gain deeper insights.

From this analysis, we can observe important trends such as the increase in content production over time, the popularity of certain genres, and variations in ratings across different types of content. We also identify frequently appearing actors and understand their contribution to the industry.

The insights gained from this project can be useful for streaming platforms and production companies. They can use this information to produce content in popular genres, improve quality, and recommend content to users based on trends and preferences.

Overall, this project demonstrates how EDA can be used to transform raw data into meaningful insights and support better decision-making in the entertainment industry.

## ***1. Know Your Data***

### Import Libraries

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

### Dataset Loading

````python
titles = pd.read_csv("/content/titles.csv.zip")
credits = pd.read_csv("/content/credits.csv.zip")
````

###Merge Dataset




````python
df = pd.merge(titles, credits, on='id')
````

### Dataset First View

````python
#Display the first 5 rows (default) for a quick preview of the dataset
df.head()
#Display the first n rows of the dataset (here, n = 10)
df.head(10)
````

### Dataset Rows & Columns count

````python
df.shape
````

### Dataset Information

````python
df.info()
````

#### Duplicate Values

````python
df.duplicated().sum()
````

#### Missing Values/Null Values

````python
df.isnull().sum()
````

````python
# Visualizing the missing values
missing = df.isnull().sum()

missing[missing > 0].plot(kind='bar')
plt.title("Missing Values Count per Column")
plt.xlabel("Columns")
plt.ylabel("Number of Missing Values")
plt.show()
````

````python
# Fill missing values in the 'description' column
# inplace=True updates the original DataFrame directly without creating a copy
# inplace=False returns a new modified copy (original data remains unchanged
df["description"].fillna("No Description", inplace=True)
df.isnull().sum()
````

````python
df["age_certification"].fillna("No certification", inplace=True)
df.isnull().sum()
````

````python
df.drop("seasons", axis=1, inplace=True)
df.isnull().sum()
````

````python
df.drop("imdb_id", axis=1, inplace=True)
df.isnull().sum()
````

````python
df["imdb_score"].fillna(df["imdb_score"].mean(), inplace=True)
df.isnull().sum()
````

````python
df["imdb_votes"].fillna(0, inplace=True)
df.isnull().sum()
````

````python
df["tmdb_popularity"].fillna(df["tmdb_popularity"].median(), inplace=True)
df.isnull().sum()
````

````python
df["tmdb_score"].fillna(df["tmdb_score"].median(), inplace=True)
df.isnull().sum()
````

````python
df["character"].fillna("No character", inplace=True)
df.isnull().sum()
````

### What did you know about your dataset?

Answer :- My dataset is a combination of two datasets: titles and credits, which were merged using a common column called id. It contains information about movies and TV shows along with cast and crew details.

The dataset includes features such as title, type (movie or show), description, release year, runtime, genres, production countries, IMDb and TMDB ratings, and popularity metrics. It also includes information about people involved, such as actor names, roles, and characters.

The dataset contains both numerical and categorical data. Some columns had missing values.

Missing values were handled based on the nature and importance of each column. Columns with a very high number of missing values, such as seasons and imdb_id, were dropped as they were not useful for analysis. Numerical columns like IMDb score and TMDB score were filled using mean values, while columns like IMDb votes were filled with 0 as missing values indicate no votes. Categorical columns such as age certification and character were filled with “Unknown”. Small missing values in description and TMDB popularity were handled using default values and median respectively to maintain data consistency.

Overall, the dataset is useful for analyzing content trends, ratings, popularity, and understanding audience preferences in movies and TV shows.
